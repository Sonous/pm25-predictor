```python
import os
import gzip
import pandas as pd

def combine_raw_data_to_csv_for_each_location(base_location_dirs: list[str]):
    for base_location_dir in base_location_dirs:
        print(f"\nProcessing location: {base_location_dir}")
        all_year_dfs = []  # Dùng để lưu toàn bộ dữ liệu của location này

        for year_folder in sorted(os.listdir(base_location_dir)):
            year_path = os.path.join(base_location_dir, year_folder)
            if not os.path.isdir(year_path):
                continue

            all_dfs = []

            # Duyệt tháng theo thứ tự
            for month_folder in sorted(os.listdir(year_path)):
                month_path = os.path.join(year_path, month_folder)
                if not os.path.isdir(month_path):
                    continue

                # Duyệt file .csv.gz theo thứ tự
                for file_name in sorted(os.listdir(month_path)):
                    if file_name.endswith(".csv.gz"):
                        file_path = os.path.join(month_path, file_name)

                        with gzip.open(file_path, 'rt', encoding='utf-8') as f:
                            df = pd.read_csv(f)
                            all_dfs.append(df)

            # Gộp dữ liệu theo năm
            if all_dfs:
                combined_df = pd.concat(all_dfs, ignore_index=True)
                output_file_year = os.path.join(base_location_dir, f"openaq_{year_folder}.csv")
                combined_df.to_csv(output_file_year, index=False)
                print(f"Saved combined CSV for {year_folder} -> {output_file_year}")

                # Lưu vào danh sách tổng cho location
                all_year_dfs.append(combined_df)

        # Gộp tất cả các năm lại cho location này
        if all_year_dfs:
            combined_location_df = pd.concat(all_year_dfs, ignore_index=True)
            output_file_location = os.path.join(base_location_dir, f"openaq_{os.path.basename(base_location_dir)}.csv")
            combined_location_df.to_csv(output_file_location, index=False)
            print(f"\n✅ Saved combined CSV for all years -> {output_file_location}")


# Prepare location dirs
base_location_dirs = []
for job in jobs:
    location_id = job["location_id"]
    location_name = job["location_name"]
    base_location_dirs.append(f"data/{location_name}_{location_id}")

# Call function
combine_raw_data_to_csv_for_each_location(base_location_dirs)
```

```python
import requests
import pandas as pd
from typing import List

def fetch_and_save_weather_data(
    latitude: float,
    longitude: float,
    start_date: str,
    end_date: str,
    timezone: str = "Asia/Singapore",
    hourly_params: List[str] = None,
    output_path: str = "weather_data.csv"
):
    """
    Lấy dữ liệu thời tiết lịch sử từ Open-Meteo API và lưu ra file CSV.

    Args:
        latitude (float): Vĩ độ.
        longitude (float): Kinh độ.
        start_date (str): Ngày bắt đầu (yyyy-mm-dd).
        end_date (str): Ngày kết thúc (yyyy-mm-dd).
        timezone (str): Múi giờ (ví dụ 'Asia/Singapore' hoặc '+08:00').
        hourly_params (List[str]): Danh sách các chỉ số thời tiết muốn lấy.
        output_path (str): Đường dẫn file CSV đầu ra.
    """
    if hourly_params is None:
        hourly_params = ["temperature_2m", "relative_humidity_2m", "precipitation", "windspeed_10m"]

    url = "https://archive-api.open-meteo.com/v1/archive"

    params = {
        "latitude": latitude,
        "longitude": longitude,
        "start_date": start_date,
        "end_date": end_date,
        "hourly": hourly_params,
        "timezone": timezone
    }

    print(f"📡 Fetching data from {start_date} to {end_date} for ({latitude}, {longitude})...")
    response = requests.get(url, params=params)
    response.raise_for_status()
    data = response.json()

    # Kiểm tra dữ liệu trả về
    if "hourly" not in data:
        print("⚠️ Không có dữ liệu được trả về.")
        return

    # Chuyển sang DataFrame
    df = pd.DataFrame(data["hourly"])
    print(f"✅ Lấy thành công {len(df)} dòng dữ liệu.")

    # Lưu ra CSV
    df.to_csv(output_path, index=False)
    print(f"💾 Đã lưu dữ liệu ra file: {output_path}")

    return df


START_DATE = "2022-11-01"
END_DATE = "2025-09-30"
TIMEZONE = "Asia/Singapore"
HOURLY = ["temperature_2m", "relative_humidity_2m", "wind_speed_10m", "wind_direction_10m", "surface_pressure", "precipitation"]

weather_jobs = [
    {
        "latitude": 22.2888888888888,
        "longitude": 113.943611111111,
        "output_path": "data/weather_location_7727.csv"
    },
    {
        "latitude": 22.4452777777777,
        "longitude": 114.022777777777,
        "output_path": "data/weather_location_7739.csv"
    },
    {
        "latitude": 22.3911111111111,
        "longitude": 113.976666666666,
        "output_path": "data/weather_location_7742.csv"
    },
    {
        "latitude": 22.49671,
        "longitude": 114.12824,
        "output_path": "data/weather_location_233335.csv"
    },
    {
        "latitude": 22.24743,
        "longitude": 114.16015,
        "output_path": "data/weather_location_233336.csv"
    },
    {
        "latitude": 22.28,
        "longitude": 114.185,
        "output_path": "data/weather_location_7732.csv"
    },
    {
        "latitude": 22.285,
        "longitude": 114.144444444444,
        "output_path": "data/weather_location_7730.csv"
    },
    {
        "latitude": 22.3133333333333,
        "longitude": 114.224722222222,
        "output_path": "data/weather_location_7735.csv"
    },
    {
        "latitude": 22.4508333333333,
        "longitude": 114.164444444444,
        "output_path": "data/weather_location_7737.csv"
    },
    {
        "latitude": 22.3225,
        "longitude": 114.1683333,
        "output_path": "data/weather_location_7728.csv"
    },
    {
        "latitude": 22.37638889,
        "longitude": 114.1844444,
        "output_path": "data/weather_location_7733.csv"
    },
    {
        "latitude": 22.33027778,
        "longitude": 114.1591667,
        "output_path": "data/weather_location_7734.csv"
    },
    {
        "latitude": 22.35722222,
        "longitude": 114.1297222,
        "output_path": "data/weather_location_7736.csv"
    },
   {
        "latitude": 22.37166667,
        "longitude": 114.1144444,
        "output_path": "data/weather_location_7740.csv"
    },
]

for job in weather_jobs:
  fetch_and_save_weather_data(
      latitude=job["latitude"],
      longitude=job["longitude"],
      start_date=START_DATE,
      end_date=END_DATE,
      timezone=TIMEZONE,
      hourly_params=HOURLY,
      output_path=job["output_path"]
  )
```
